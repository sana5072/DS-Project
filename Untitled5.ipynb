```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Disaster_Type'] = label_encoder.fit_transform(df['Disaster_Type'])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df["Location"] = label_encoder.fit_transform(df['Location'])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df["Date"] = label_encoder.fit_transform(df["Date"])
```

```python
df.head()
```

```python
# Drop Disaster id column as it is not needed
df.drop(columns=["Disaster_ID"],inplace=True)
```

```python
df.head()
```

```python
# seperate x and y variable

x = df.drop(columns=["Economic_Loss($)"])  # Features
y = df["Economic_Loss($)"]  # Target variable
```

```python
x
```

```python
# Step 4: Split data into training and testing sets (80% train, 20% test)
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
# Step 8: Evaluate the model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(ytest, ypred)
mse = mean_squared_error(ytest, ypred)
r2 = r2_score(ytest, ypred)

# Print results
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```
